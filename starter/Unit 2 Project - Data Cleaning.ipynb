```python
### DEFINE ###

    # Change to snake case after removing punctuation and spaces from column names.

### CLEAN ### 

import regex as re

column_list = house_df.columns.to_list()
new_columns = []

for column in column_list:
    column = column.replace(" ", "_")
    column = column.replace("?", "")
        
    new_columns.append(column)

house_df.columns = new_columns

### TEST ###

house_df.head()
```

```python
### DEFINE ###

    # Change id column to a string from integer. 

### CLEAN ### 

house_df.id = house_df.id.astype(str)

### TEST ###

print(house_df['id'])
```

```python
### DEFINE ###

    # change date column to datetime data type from string

### CLEAN ### 

sales_df['date'] = sales_df['date'].str.replace('T000000', '') 

sales_df['date'] = pd.to_datetime(sales_df['date'], format='%Y%m%d')

### TEST ###

print(sales_df['date'])
```
